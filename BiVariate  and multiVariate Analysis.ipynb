### Scatterplot  numerical x numerical

```python
sns.scatterplot(data=tips, x='total_bill', y='tip')
```

```python
# Multivariate
sns.scatterplot(data=tips, x='total_bill', y='tip', hue= 'sex')
```

```python
#Multivariate 4 columns 
sns.scatterplot(data=tips, x='total_bill', y='tip', hue= 'sex', style = 'smoker')
```

```python
#Multivariate 5 columns 
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex', style='smoker', size='size')
```

### Distplot- Numerical x Categorical

### Heat Map Categorical x Categorical

```python
titanic.head(2)
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
titanic.groupby('Pclass').mean()['Survived']*100
```

```python
(titanic.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

```python
(titanic.groupby('Sex').mean()['Survived']*100)
```

```python
(titanic.groupby('Embarked').mean()['Survived']*100)
```
